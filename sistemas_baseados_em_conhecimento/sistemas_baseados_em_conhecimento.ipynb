## Funções Auxiliares

```python
#Função de validação da inserção do bioma
def validarBioma():

  #O bioma é questionado até que uma entrada válida seja inserida pelo usuário    
  while True:

    bioma = input()

    if bioma in ["Floresta Amazônica", "Mata Atlântica", "Cerrado", "Caatinga", "Pantanal", "Pampa"]:

      break
          
    print("Opção inválida. Tente novamente.")

  return bioma
```

```python
#Função de validação da inserção do clima
def validarClima():

  #O clima é questionado até que uma entrada válida seja inserida pelo usuário    
  while True:

    clima = input()

    if clima in ["Tropical", "Subtropical", "Semi-árido", "Equatorial"]:

      break
      
    print("Opção inválida. Tente novamente.")
    
  return clima
```

```python
#Função de validação da inserção de perguntas de sim ou não 
def validarEntrada():

  while True:

    entrada = input()

    #O usuário é questionado até que uma entrada válida seja inserida     
    if entrada in ["S", "N"]:

      break
      
    print("Opção inválida. Tente novamente.")
        
  return entrada
```

## Implementação com a Biblioteca

```python
#Definição da classe de identificação de Estados
class IdentificarEstados(KnowledgeEngine):
  
  #Inicialmente, a identificação não possui erros
  @DefFacts()
  def _initial_action(self):
        
    yield Fact(erro = "N")

  #Identificação do Estado RS
  @Rule(Fact(bioma = "Pampa"))
  def estadoRS(self):

    print("O Estado é Rio Grande do Sul.")

  #Identificação do Estado MS
  @Rule(Fact(bioma = "Pantanal"))
  def estadoMS(self):

    print("O Estado é Mato Grosso do Sul.")

  #Identificação do Estado MT
  @Rule(Fact(bioma = "Floresta Amazônica"), 
        Fact(clima = "Tropical"))
  def estadoMT(self):

    print("O Estado é Mato Grosso.")

  #Identificação do Estado PA
  @Rule(Fact(bioma = "Floresta Amazônica"), 
        Fact(clima = "Equatorial"), 
        Fact(habitantes = "S"))
  def estadoPA(self):

    print("O Estado é Pará.")

  #Identificação do Estado TO
  @Rule(Fact(bioma = "Cerrado"), 
        Fact(habitantes = "N"))
  def estadoTO(self):

    print("O Estado é Tocantins.")

  #Identificação do Estado PI
  @Rule(Fact(bioma = "Caatinga"), 
        Fact(clima = "Tropical"), 
        Fact(habitantes = "N"))
  def estadoPI(self):

    print("O Estado é Piauí.")

  #Identificação do Estado PE
  @Rule(Fact(bioma = "Caatinga"), 
        Fact(clima = "Semi-árido"), 
        Fact(habitantes = "S"))
  def estadoPE(self):

    print("O Estado é Pernambuco.")

  #Identificação do Estado ES
  @Rule(Fact(bioma = "Mata Atlântica"), 
        Fact(clima = "Tropical"), 
        Fact(habitantes = "N"))
  def estadoES(self):

    print("O estado é Espírito Santo.")

  #Identificação do Estado SP
  @Rule(Fact(bioma = "Mata Atlântica"), 
        Fact(clima = "Tropical"), 
        Fact(habitantes = "S"), 
        Fact(area = "S"))
  def estadoSP(self):

    print("O Estado é São Paulo.")

  #Identificação do Estado RJ
  @Rule(Fact(bioma = "Mata Atlântica"), 
        Fact(clima = "Tropical"), 
        Fact(habitantes = "S"), 
        Fact(area = "N"))
  def estadoRJ(self):

    print("O Estado é Rio de Janeiro.")

  #Identificação do Estado PR
  @Rule(Fact(bioma = "Mata Atlântica"), 
        Fact(clima = "Subtropical"),
        Fact(area = "S"))
  def estadoPR(self):

    print("O Estado é Paraná.")

  #Identificação do Estado SC
  @Rule(Fact(bioma = "Mata Atlântica"), 
        Fact(clima = "Subtropical"), 
        Fact(area = "N"))
  def estadoSC(self):

    print("O Estado é Santa Catarina.")

  #Identificação do Estado BA
  @Rule(Fact(bioma = "Caatinga"), 
        Fact(clima = "Tropical"), 
        Fact(habitantes = "S"), 
        Fact(area = "S"))
  def estadoBA(self):

    print("O Estado é Bahia.")

  #Identificação do Estado CE
  @Rule(Fact(bioma = "Caatinga"), 
        Fact(clima = "Tropical"), 
        Fact(habitantes = "S"), 
        Fact(area = "N"))
  def estadoCE(self):

    print("O Estado é Ceará.")

  #Identificação do Estado RR
  @Rule(Fact(bioma = "Floresta Amazônica"), 
        Fact(clima = "Equatorial"), 
        Fact(habitantes = "N"), 
        Fact(area = "S"), 
        Fact(idh = "S"))
  def estadoRR(self):

    print("O Estado é Roraima.")

  #Identificação do Estado AP
  @Rule(Fact(bioma = "Floresta Amazônica"), 
        Fact(clima = "Equatorial"), 
        Fact(habitantes = "N"), 
        Fact(area = "N"), 
        Fact(idh = "S"))
  def estadoAP(self):

    print("O Estado é Amapá.")

  #Identificação do Estado AC
  @Rule(Fact(bioma = "Floresta Amazônica"), 
        Fact(clima = "Equatorial"), 
        Fact(habitantes = "N"), 
        Fact(area = "N"), 
        Fact(idh = "N"))
  def estadoAC(self):

    print("O Estado é Acre.")

  #Identificação do Estado AM
  @Rule(Fact(bioma = "Floresta Amazônica"), 
        Fact(clima = "Equatorial"), 
        Fact(habitantes = "N"), 
        Fact(area = "S"), 
        Fact(idh = "N"), 
        Fact(pib = "S"))
  def estadoAM(self):

    print("O Estado é Amazonas.")

  #Identificação do Estado RO
  @Rule(Fact(bioma = "Floresta Amazônica"), 
        Fact(clima = "Equatorial"), 
        Fact(habitantes = "N"), 
        Fact(area = "S"), 
        Fact(idh = "N"), 
        Fact(pib = "N"))
  def estadoRO(self):

    print("O Estado é Rondônia.")

  #Identificação do Estado RN
  @Rule(Fact(bioma = "Caatinga"), 
        Fact(clima = "Semi-árido"), 
        Fact(habitantes = "N"), 
        Fact(pib = "S"))
  def estadoRN(self):

    print("O Estado é Rio Grande do Norte.")

  #Identificação do Estado PB
  @Rule(Fact(bioma = "Caatinga"), 
        Fact(clima = "Semi-árido"), 
        Fact(habitantes = "N"), 
        Fact(pib = "N"))
  def estadoPB(self):

    print("O Estado é Paraíba.")

  #Identificação do Estado MG
  @Rule(Fact(bioma = "Cerrado"), 
        Fact(habitantes = "S"), 
        Fact(idh = "S"), 
        Fact(vida = "S"))
  def estadoMG(self):

    print("O Estado é Minas Gerais.")

  #Identificação do Estado GO
  @Rule(Fact(bioma = "Cerrado"), 
        Fact(habitantes = "S"), 
        Fact(idh = "S"), 
        Fact(vida = "N"))
  def estadoGO(self):

    print("O Estado é Goiás.")

  #Identificação do Estado MA
  @Rule(Fact(bioma = "Cerrado"), 
        Fact(habitantes = "S"), 
        Fact(idh = "N"))
  def estadoMA(self):

    print("O Estado é Maranhão.")

  #Identificação do Estado AL
  @Rule(Fact(bioma = "Mata Atlântica"), 
        Fact(clima = "Semi-árido"), 
        Fact(municipios = "S"))
  def estadoAL(self):

    print("O Estado é Alagoas.")

  #Identificação do Estado SE
  @Rule(Fact(bioma = "Mata Atlântica"), 
        Fact(clima = "Semi-árido"), 
        Fact(municipios = "N"))
  def estadoSE(self):

    print("O Estado é Sergipe.")

  #Questionamento sobre o bioma predominante do Estado
  @Rule(Fact(erro = "N"),
        NOT(Fact(bioma = W())))
  def perguntarBioma(self):

    print("Qual é o bioma predominante do Estado? (Floresta Amazônica, Mata Atlântica, Cerrado, Caatinga, Pantanal, Pampa)")
    self.declare(Fact(bioma = validarBioma()))

  #Questionamento sobre o clima predominante do Estado
  @Rule(Fact(erro = "N"),
        NOT(Fact(clima = W())), 
        Fact(bioma = W()),
        NOT(Fact(bioma = "Pampa")),
        NOT(Fact(bioma = "Pantanal")), 
        NOT(Fact(bioma = "Cerrado")))
  def perguntarClima(self):

    print("Qual é o clima predominante do Estado? (Tropical, Subtropical, Semi-árido, Equatorial)")
    self.declare(Fact(clima = validarClima()))
      
  #Questionamento sobre a quantidade de habitantes do Estado
  @Rule(Fact(erro = "N"),
        NOT(Fact(habitantes = W())), 
        OR(Fact(bioma = "Cerrado"), 
           AND(Fact(bioma = "Mata Atlântica"), Fact(clima = "Tropical")), 
           AND(Fact(bioma = "Floresta Amazônica"), Fact(clima = "Equatorial")), 
           AND(Fact(bioma = "Caatinga"), Fact(clima = "Tropical")), 
           AND(Fact(bioma = "Caatinga"), Fact(clima = "Semi-árido"))))    
  def perguntarHabitantes(self):

    print("A população do Estado é maior que 5 milhões de habitantes? (S/N)")
    self.declare(Fact(habitantes = validarEntrada()))

  #Questionamento sobre a área do Estado
  @Rule(Fact(erro = "N"),
          NOT(Fact(area = W())), 
          OR(AND(Fact(bioma = "Mata Atlântica"), Fact(clima = "Subtropical")), 
             AND(Fact(bioma = "Mata Atlântica"), Fact(clima = "Tropical"), Fact(habitantes = "S")), 
             AND(Fact(bioma = "Floresta Amazônica"), Fact(clima = "Equatorial"), Fact(habitantes = "N")), 
             AND(Fact(bioma = "Caatinga"), Fact(clima = "Tropical"), Fact(habitantes = "S"))))
  def perguntarArea(self):

    print("A área do Estado é maior que 170 milhões de KM^2? (S/N)")
    self.declare(Fact(area = validarEntrada()))

  #Questionamento sobre o IDH do Estado
  @Rule(Fact(erro = "N"),
        NOT(Fact(idh = W())), 
        OR(AND(Fact(bioma = "Cerrado"), Fact(habitantes = "S")), 
           AND(Fact(bioma = "Floresta Amazônica"), Fact(clima = "Equatorial"), Fact(habitantes = "N"))))
  def perguntarIDH(self):

    print("O IDH do Estado é maior que 0.735? (S/N)")
    self.declare(Fact(idh = validarEntrada()))

  #Questionamento sobre o PIB do Estado
  @Rule(Fact(erro = "N"),
        NOT(Fact(pib = W())), 
        OR(AND(Fact(bioma = "Floresta Amazônica"), Fact(clima = "Equatorial"), Fact(habitantes = "N"), Fact(area = "S"), Fact(idh = "N")), 
           AND(Fact(bioma = "Caatinga"), Fact(clima = "Semi-árido"), Fact(habitantes = "N"))))
  def perguntarPIB(self):

    print("O PIB do Estado é maior que R$65 bilhões? (S/N)")
    self.declare(Fact(pib = validarEntrada()))

  #Questionamento sobre a expectativa de vida do Estado
  @Rule(Fact(erro = "N"),
        NOT(Fact(vida = W())), 
        AND(Fact(bioma = "Cerrado"), Fact(habitantes = "S"), Fact(idh = "S")))
  def perguntarVida(self):

    print("A expectativa de vida do Estado é maior que 75 anos? (S/N)")
    self.declare(Fact(vida = validarEntrada()))

  #Questionamento sobre a quantidade de municípios do Estado
  @Rule(Fact(erro = "N"),
        NOT(Fact(municipios = W())), 
        AND(Fact(bioma = "Mata Atlântica"), Fact(clima = "Semi-árido")))
  def perguntarMunicipios(self):

    print("A quantidade de municípios do Estado é maior que 100? (S/N)")    
    self.declare(Fact(municipios = validarEntrada()))

  #Se houverem respostas incoerentes, isto é, um conjunto de características que não identifique nenhum
  #Estado, então um erro é retornado e o processo de identificação é encerrado
  @Rule(OR(AND(Fact(bioma = "Caatinga"), NOT(Fact(clima = "Tropical")), NOT(Fact(clima = "Semi-árido"))),
           AND(Fact(bioma = "Floresta Amazônica"), NOT(Fact(clima = "Tropical")), NOT(Fact(clima = "Equatorial"))),
           AND(Fact(bioma = "Mata Atlântica"), NOT(Fact(clima = "Tropical")), NOT(Fact(clima = "Semi-árido")), NOT(Fact(clima = "Subtropical")))))
  def erro(self):
    
    self.declare(Fact(erro = "S"))
    print("Não existe Estado com as características inseridas.")
```

## Implementação sem a Biblioteca

```python
#Definição da função de identificação de Estados
def identificarEstados():

  print("Qual é o bioma predominante do Estado? (Floresta Amazônica, Mata Atlântica, Cerrado, Caatinga, Pantanal, Pampa)")
  bioma = validarBioma()

  if bioma == "Floresta Amazônica":

    print("Qual é o clima predominante do Estado? (Tropical, Subtropical, Semi-árido, Equatorial)")
    clima = validarClima()

    if clima == "Tropical":

      print("O Estado é Mato Grosso.")
  
    elif clima == "Equatorial":

      print("A população do Estado é maior que 5 milhões de habitantes? (S/N)")
      habitantes = validarEntrada()

      if habitantes == "S":

        print("O Estado é Pará.")
    
      else:
    
        print("A área do Estado é maior que 170 milhões de KM^2? (S/N)")
        area = validarEntrada()

        print("O IDH do Estado é maior que 0.735? (S/N)")
        idh = validarEntrada()

        if area == "S":

          if idh == "S":

            print("O Estado é Roraima.")
        
          else:
        
            print("O PIB do Estado é maior que R$65 bilhões? (S/N)")
            pib = validarEntrada()

            if pib == "S":

              print("O Estado é Amazonas.")
          
            else:

              print("O Estado é Rondônia.")
          
        else:

          if idh == "S":

            print("O Estado é Amapá.")
        
          else:

            print("O Estado é Acre.")

    else:
    
      print("Não existe Estado com as características inseridas.")

  elif bioma == "Mata Atlântica":

    print("Qual é o clima predominante do Estado? (Tropical, Subtropical, Semi-árido, Equatorial)")
    clima = validarClima()

    if clima == "Tropical":
    
      print("A população do Estado é maior que 5 milhões de habitantes? (S/N)")
      habitantes = validarEntrada()

      if habitantes == "S":

        print("A área do Estado é maior que 170 milhões de KM^2? (S/N)")
        area = validarEntrada()

        if area == "S":

          print("O Estado é São Paulo.")
      
        else:

          print("O Estado é Rio de Janeiro.")
      
      else:

        print("O Estado é Espírito Santo.")

    elif clima == "Subtropical":

      print("A área do Estado é maior que 170 milhões de KM^2? (S/N)")
      area = validarEntrada()

      if area == "S":

        print("O Estado é Paraná.")
    
      else:

        print("O Estado é Santa Catarina.")
    
    elif clima == "Semi-árido":
    
      print("A quantidade de municípios do Estado é maior que 100? (S/N)")
      municipios = validarEntrada()

      if municipios == "S":

        print("O Estado é Alagoas.")
    
      else:

        print("O Estado é Sergipe.")

    else:
  
      print("Não existe Estado com as características inseridas.")

  elif bioma == "Cerrado":

    print("A população do Estado é maior que 5 milhões de habitantes? (S/N)")
    habitantes = validarEntrada()

    if habitantes == "S":

      print("O IDH do Estado é maior que 0.735? (S/N)")
      idh = validarEntrada()

      if idh == "S":

        print("A expectativa de vida do Estado é maior que 75 anos? (S/N)")
        vida = validarEntrada()

        if vida == "S":

          print("O Estado é Minas Gerais.")
      
        else:

          print("O Estado é Goiás.")
    
      else:

        print("O Estado é Maranhão.")
    
    else:

      print("O Estado é Tocantins.")
  
  elif bioma == "Caatinga":

    print("Qual é o clima predominante do Estado? (Tropical, Subtropical, Semi-árido, Equatorial)")
    clima = validarClima()

    if clima == "Tropical":

      print("A população do Estado é maior que 5 milhões de habitantes? (S/N)")
      habitantes = validarEntrada()

      if habitantes == "S":

        print("A área do Estado é maior que 170 milhões de KM^2? (S/N)")
        area = validarEntrada()

        if area == "S":

          print("O Estado é Bahia.")
      
        else:

          print("O Estado é Ceará.")
    
      else:

        print("O Estado é Piauí.")
  
    elif clima == "Semi-árido":

      print("A população do Estado é maior que 5 milhões de habitantes? (S/N)")
      habitantes = validarEntrada()

      if habitantes == "S":

        print("O Estado é Pernambuco.")
    
      else:

        print("O PIB do Estado é maior que R$65 bilhões? (S/N)")
        pib = validarEntrada()

        if pib == "S":

          print("O Estado é Rio Grande do Norte.")
      
        else:

          print("O Estado é Paraíba.")

    else:

      print("Não existe Estado com as características inseridas.")

  elif bioma == "Pantanal":

    print("O Estado é Mato Grosso do Sul.")

  else:

    print("O Estado é Rio Grande do Sul.")
```

## Função Principal

```python
#Função principal para a realização da identificação de Estados, com ambas as implementações
def principal(opcao):

  print("Deseja identificar um Estado do Brasil (S/N)?")
  entrada = validarEntrada()

  if entrada == "S":

    while True:

      #Identificação realizada pela implementação com a biblioteca
      if opcao == 1:

        engine = IdentificarEstados()
        engine.reset()
        engine.run()
      
      #Identificação realizada pela implementação sem a biblioteca
      else:

        identificarEstados()

      print("Deseja identificar outro Estado do Brasil (S/N)?")
      entrada = validarEntrada()

      if entrada == "N":

        print("Tchau.")
        break

  else:

    print("Tchau.")
```
